## Get Top Student by Total Score

Develop a function to get the top student by total score. This is primarily to test the ability to manipulate Spark Data Frames, sort the data and get the top record.
* We will provide Spark Data Frame with student id followed by scores for different subjects.
* Develop a function which takes students as input and return the topper with highest total score.

```python
students = [(1, 72, 72, 74), (2, 69, 90, 88), (3, 90, 95, 93), (4, 47, 47, 44), (5, 76, 78, 75), (6, 71, 83, 78)]

students_df = spark.createDataFrame(students, schema='student_id INT, math INT, social INT, science INT')
```

### Step 2: Provide the solution

Now come up with the solution by developing the required logic. Once the function is developed, go to the next step to take care of the validation.

**Hint: Sort the data by total score in descending order and use `first` to get the highest scoring student details.**

```python
from pyspark.sql.functions import col
def get_topper_by_total_score(students_df):
    ## your code should go here
    topper_by_total_score = students_df. \
        withColumn('total_score', col('math') + col('social') + col('science')). \
        orderBy(col('total_score').desc()). \
        first()
    return topper_by_total_score
```

### Step 3: Validate the function

Let us validate the function by running below cells.
* The student id with highest total score is **3**.

```python
topper_by_total_score = get_topper_by_total_score(students_df)
```

```python
topper_by_total_score['student_id'] # 3
```
